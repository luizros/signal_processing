```python
import matplotlib.pyplot as plt
import numpy as np
```

### Função impulso

A função impulso pode ser definida como

$$

```python
def impulso(t):
    N = np.zeros(shape = (len(t), ))
    for k in range(len(t)):
        if t[k] == 0:
            N[k] = 1 
        else:
            N[k] = 0
    return N
```

```python
def degrau(t):
    N = np.zeros(shape = (len(t), ))
    for k in range(len(t)):
        if t[k] >= 0:
            N[k] = 1 
        else:
            N[k] = 0
    return N
```

```python
# Exemplo do impulso unitário em tempo discreto
n = np.arange(-10,10,)
plt.stem(n, impulso(n), basefmt="r")
plt.xticks(list(range(-10,11,1))) 
plt.ylabel("$\delta(t)$")
plt.xlabel("$n$")
plt.grid()
plt.show()
```

```python
# Exemplo do degrau unitário em tempo discreto
n = np.arange(-10,10,1)
plt.stem(n,degrau(n), basefmt="r")
plt.xticks(list(range(-10,11,1))) 
plt.ylabel("$\delta(t)$")
plt.xlabel("$n$")
plt.grid()
plt.show()
```

```python
# Exemplo do degrau unitário em tempo contínuo
n = np.arange(-10,10,0.001)
plt.plot(n, degrau(n))
plt.xticks(list(range(-10,11,1)))    # changing x scale by own
plt.yticks(list(range(-1,2,1)))    # changing x scale by own
plt.ylabel("$\delta(t)$")
plt.xlabel("$n$")
plt.grid(True)
plt.show()
```

```python
# Exemplo de exercício E.18 - SINAIS E SISTEMAS LINEARES BY LATHI

t = np.arange(-4,8,0.001)

def x(t):
    return (t-1)*degrau(t-1)-(t-2)*degrau(t-2)-degrau(t-4)

plt.plot(t, x(t),linewidth=2, label=r"x(t) = $(t-1)u(t-1)-(t-2)u(t-2)-u(t-4)$")
plt.xticks(list(range(-4,8,1)))    # changing x scale by own
plt.yticks(list(range(-1,2,1)))    # changing x scale by own
plt.axhline(y=1.0, color="black", linestyle="--")
plt.axhline(y=-1.0, color="black", linestyle="--")
plt.ylabel("$x(t)$")
plt.xlabel("$t$")
plt.legend(fontsize=8,loc=4)
plt.grid(True)
plt.show()
```
